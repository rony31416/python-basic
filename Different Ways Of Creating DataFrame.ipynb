## Different Ways Of Creating DataFrame

```python
import pandas as pd

df = pd.read_csv('random_weather_data.csv')
```

```python
df
```

```python
#for excel
#df = pd.read_excel('the_excel_sheetname','the page of the excel such Sheet1')

#Another way is using dataFrame or Python Dictionary 

data = {
    'name' : ['Rony','Rahim','Karin'],
    'age' : ['20','30','24']
}

df = pd.DataFrame(data)
df
```

```python
#Touple list

data = (
    ("Rony", 20),
    ("Rahim", 30),
    ("Karin", 24)
)

df = pd.DataFrame(data,columns=['name','age'])
df
```

```python
data = [
    {"name": "Rony", "age": 20},
    {"name": "Rahim", "age": 30},
    {"name": "Karin", "age": 24}
]


df = pd.DataFrame(data)
df
```
